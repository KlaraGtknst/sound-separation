```python
import os
import numpy as np
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
os.chdir("../")
```

```python
ds_name = "XCM"
cache_dir = f"/mnt/work/bird2vec/logs_mw/data_gadme/{ds_name}"
```

```python
ds = load_dataset("DBD-research-group/BirdSet", ds_name, cache_dir=cache_dir, trust_remote_code=True, split="train")
```

```python
lens = ds["length"]
uniques, counts = np.unique(lens, return_counts=True)
plt.bar(uniques, height=counts)
plt.title(f"Total size: {len(lens)}")
```

```python
ds = ds.filter(lambda x: 120 >= x["length"])
```

```python
lens = ds["length"]
uniques, counts = np.unique(lens, return_counts=True)
plt.bar(uniques, height=counts)
plt.axvline(120, color="r", label=f"Size_120: {len(lens)}")
plt.axvline(60, color="g", label=f"Size_60: {len([i for i in lens if i <= 60])}")
plt.axvline(30, color="b", label=f"Size_30: {len([i for i in lens if i <= 30])}")
plt.axvline(20, color="y", label=f"Size_20: {len([i for i in lens if i <= 20])}")
plt.xlabel("length <= 120")
plt.legend()
```

```python
ds = ds.filter(lambda x: 5 <= x["length"] <= 120)
```

```python
lens = ds["length"]
uniques, counts = np.unique(lens, return_counts=True)
plt.bar(uniques, height=counts)
plt.xlabel("5 <= length <= 120")
plt.axvline(120, color="r", label=f"Size_120: {len(lens)}")
plt.axvline(60, color="g", label=f"Size_60: {len([i for i in lens if i <= 60])}")
plt.axvline(30, color="b", label=f"Size_30: {len([i for i in lens if i <= 30])}")
plt.axvline(20, color="y", label=f"Size_20: {len([i for i in lens if i <= 20])}")
plt.legend()
```
